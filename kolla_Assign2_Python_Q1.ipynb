# we use try and except so as some countrys are islands and they are officially not recognized.

```python
import pycountry_convert as pc
for Name in range(df.Country.count()):
 try:
    country_code = pc.country_name_to_country_alpha2(df.Country[Name])
    continent_name = pc.country_alpha2_to_continent_code(country_code)
    df1 = continent_name
    #df.iloc[Name]=({'Continent':df1})
    #df['Continent'] = df1
#df.at[Name,'Continent'] = df1[Name]
    df.loc[df.index[Name], 'Continent'] = df1
 except KeyError:df.at[Name, 'Continent'] = 'NaN'

    #df.insert(1,'Continent','NaN')
    #df.insert(1,'Continent', continent_name)
```

# I used replace fuction to replace the Continent codes with their Names.

```python
df['Continent']=df['Continent'].replace('AS','ASIA').replace('NA','NORTH AMERICA').replace('SA','SOUTH AMERICA').replace('OC','OCEANIA').replace('AN','ANTARTICA').replace('AF','AFRICA').replace('EU','EUROPE').replace('NaN','OTHER')
```
